# Loading the NBA Dataset

Let's read in the Games dataset. This dataset contains the performance of each player in each game from the 2003 to 2019 season. It is a CSV file so we will use `.read_csv()`.

We will now use `.shape()` to and `.head()` to get a better idea of what our data looks like. We can use `.dtypes` to check our column headers and data types.

```python
games = pd.read_csv(NBA_GAMES_DATASET)
games.shape, games.dtypes
```

We want the column `GAME_DATE_EST` to be of type datetime.

```python
games = pd.read_csv(NBA_GAMES_DATASET, parse_dates=['GAME_DATE_EST'])
games.shape, games.dtypes.head(5)
```

```python
games.head()
```

# Checking for Missing Data

Some rows have incomplete data.

```python
games.isnull()
```

We can use `.isnull()` in conjunction with `.sum()` to count the number of empty cells.

```python
games.isnull().sum()
```

To filter out rows that don't contain data we can use a mask. This mask will filter out all games where the home team hasn't scored any points. We know this won't remove any rows we don't want to remove because there has never been a game where a team scored 0.

```python
more_than_0_point = games['PTS_home']>0
more_than_0_point
```

```python
games_with_data = games[more_than_0_point]
games_with_data.shape
```

```python
games_with_data.isnull().sum()
```

```python
games_with_data
```

Let's make a quick visualisation of the number of points scored by home and away teams by season to check the data is how we expect it.

```python
score_by_season = games_with_data[['GAME_DATE_EST', 'SEASON', 'PTS_home', 'PTS_away']].set_index('GAME_DATE_EST').sort_index()
score_by_season.groupby(['SEASON']).mean()
```

```python
score_by_season.groupby(['SEASON']).mean().plot(kind = "bar", figsize=(13,5))
plt.ylim([80, 120])
plt.show()
```

This graph shows an obvious difference in scoring of home and away teams. Let's create another graph to take a better look at this

```python
fig, ax = plt.subplots(figsize=(13, 5))

away = score_by_season.groupby(['SEASON']).mean()['PTS_away']

home = score_by_season.groupby(['SEASON']).mean()['PTS_home']

plt.plot(score_by_season.groupby(['SEASON']).mean().index, score_by_season.groupby(['SEASON']).mean()['PTS_home'])
plt.plot(score_by_season.groupby(['SEASON']).mean().index, score_by_season.groupby(['SEASON']).mean()['PTS_away'])

ax.fill_between(score_by_season.groupby(['SEASON']).mean().index, away, home, color='steelblue', alpha=.2)

ax.legend(['PTS home', 'PTS_away'], loc='upper left', frameon=False)
plt.title('Average Home vs Away points by Season')
plt.xlabel('Season')
plt.ylabel('Points')

fig.tight_layout()
```

# Save Base Games Dataset
Now that the dataset is cleaned we can save it to our processed data folder

```python
games_with_data.to_csv(PROCESSED_GAMES_DATASET, index=False)
games_with_data.columns, PROCESSED_GAMES_DATASET
```
