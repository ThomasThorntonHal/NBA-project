# Loading the NBA Dataset

```python
games = pd.read_csv(GAMES_DATASET, usecols=range(1,32), parse_dates=['Date'])
games.shape, games['Date'].max()
```

```python
games.columns
```

```python
games.head()
```

Check for null data

```python
games.isnull().sum()
```

You can see the `Season` column is poorly formatted.

```python
games.groupby(['Season', 'Team']).mean()
```

Let's make a dictionary to correct the `Season` data

```python
seasons = {96:1996, 97:1997, 98:1998, 99:1999, 0:2000, 1:2001, 2:2002, 3:2003, 4:2004, 5:2005, 
           6:2006, 7:2007, 8:2008, 9:2009, 10:2010, 11:2011, 12:2012, 13:2013, 14:2014, 15:2015, 
           16:2016, 17:2017, 18:2018}
```

We can now map this dictionary to the dataframe.

```python
clean_games = games.copy()
clean_games['Season'] = games['Season'].map(seasons)
```

# Save Dataset
Now that the dataset has been cleaned we can save it.

```python
clean_games.to_csv(CLEAN_GAMES_DATASET, index=False)
```
