```python
player_season_stats = clean_player_game_stats.groupby(["PLAYER_ID", "SEASON"]).mean()
player_season_stats
```

```python
player_season_stats.columns
```

```python
player_season_stats['FG_PCT'] = player_season_stats['FGM']/player_season_stats['FGA']
player_season_stats['FG3_PCT'] = player_season_stats['FG3M']/player_season_stats['FG3A']
player_season_stats['FT_PCT'] = player_season_stats['FTM']/player_season_stats['FTA']
player_season_stats.head(1)
```

```python
player_season_stats['POSSESSIONS'] = round(0.96 * (player_season_stats['FGA'] + player_season_stats['TO'] + 
                                                  0.44*player_season_stats['FTA'] - player_season_stats['OREB']))
player_season_stats.head(1)
```

```python
player_season_stats['OER'] = round(100*(player_season_stats['PTS']/player_season_stats['POSSESSIONS']))
player_season_stats.head(5)
```

```python
player_season_stats.loc[1717]
```

# Efficiency

(PTS + REB + AST + STL + BLK − Missed FG − Missed FT - TO) / GP

```python
player_game_stats.columns
```

```python
player_game_stats['EFF'] = player_game_stats['PTS']+player_game_stats['DREB']+player_game_stats['OREB']+\
                            player_game_stats['AST']+player_game_stats['STL']+player_game_stats['BLK']-\
                            (player_game_stats['FGA']-player_game_stats['FGM'])-(player_game_stats['FTA']-player_game_stats['FTM'])-\
                            player_game_stats['TO']
player_game_stats
```
