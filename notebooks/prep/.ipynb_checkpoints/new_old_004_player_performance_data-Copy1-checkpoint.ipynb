# Introduction
In this notebook I will be preparing the game and player data for analysis. I will also be merging these two datasets to make a new dataset which will be used for analysis later. We need the data of the team's performance as well as the data of the individual players performance in one dataframe.

# Setup
## Imports

```python
import pandas as pd
import numpy as np
```

# Parameters

```python
# in
GAMES_DATA = 'C:/Users/User/OneDrive/Desktop/DSIP/NBAproject/data/raw/games.csv'
PLAYER_DATA = 'C:/Users/User/OneDrive/Desktop/DSIP/NBAproject/data/raw/games_details.csv'

# out
CLEAN_PLAYER_DATA = 'C:/Users/User/OneDrive/Desktop/DSIP/NBAproject/data/processed/004_player_game_data.csv'
FULL_ADVANCED_STATS = 'C:/Users/User/OneDrive/Desktop/DSIP/NBAproject/data/processed/004_advanced_data.csv'
```

# Load Datasets

## Player Dataset

```python
player_data = pd.read_csv(PLAYER_DATA)
player_data
```

```python
player_data.dtypes
```

## Games dataset

```python
game_data = pd.read_csv(GAMES_DATA)
game_data.head(5)
```

## Combine player and game data

```python
combined = pd.merge(player_data, game_data)
combined.head(5)
```

```python
# The resulting df should be no larger than the player_data.
assert len(combined)<= len(player_data)
```

```python
combined.columns
```

We can filter out the columns we don't need by making a list of the columns we want to keep.

```python
combined = combined[combined['MIN'].notnull()]
combined.isnull().sum()
```

# Save Dataset
This dataset will be used later in analysis notebook `410_player_efficiency`

```python
combined.to_csv(FULL_ADVANCED_STATS, index=False)
```

# Continue with player data

```python
columns = ["SEASON","GAME_ID","TEAM_ID","PLAYER_ID","PLAYER_NAME","TEAM_ABBREVIATION","START_POSITION","MIN","FGM","FGA","FG_PCT",
           "FG3M","FG3A","FG3_PCT","FTM","FTA","FT_PCT","OREB","DREB","REB","AST","STL","BLK","TO","PF","PTS","PLUS_MINUS"]
```

```python
player_game_stats = combined[columns]
player_game_stats
```

```python
player_game_stats.isnull().sum()
```

Check for null rows

```python
player_game_stats[player_game_stats['MIN'].isnull()].groupby(['PLAYER_ID', 'PLAYER_NAME']).sum()
```

```python
clean_player_game_stats = player_game_stats[player_game_stats['MIN'].notnull()]
clean_player_game_stats.isnull().sum()
```

# Save Dataset
This dataset will also be used in `410_player_efficiency`

```python
clean_player_game_stats.to_csv(CLEAN_PLAYER_DATA, index=False)
CLEAN_PLAYER_DATA
```

```python
del player_data
del game_data
del combined
del player_game_stats
del clean_player_game_stats
```
