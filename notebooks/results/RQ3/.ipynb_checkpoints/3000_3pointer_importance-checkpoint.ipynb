# Introduction
In this notebook we will look at how important the 3-point shot is

# Notebook Objectives
We are interested in two questions:
1. Is the 3-point shot important for a team's winning chances?
2. Do attempts or percentage made matter more?

## Imports

```python
import pandas as pd

import numpy as np

import matplotlib.pylab as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

import matplotlib

import seaborn as sns

import numpy as np
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import LinearSegmentedColormap
from scipy import misc
import nba_api as nba

%matplotlib inline
```

```python
SHOT_DATA = '../../../data/processed/310_3point_opp.csv'
```

## Configuration

```python
sns.set_context('talk')

%matplotlib inline
```

## Load shots dataset

```python
df = pd.read_csv(SHOT_DATA)
df.shape, df.nunique()
```

# Does better 3-point shooting matter?

## 2018

```python
fig, ax = plt.subplots(figsize=(12,11))

x = df[(df['Season'] == 2018) & (df['Won'] == 1)]
y = df[(df['Season'] == 2018) & (df['Won'] == 0)]

scatter = ax.scatter(x['comp_3PM'], x['comp_3P%'], color='green', alpha=0.26, s=abs(x['comp_Total'])**1.8)
ax.scatter(y['comp_3PM'], y['comp_3P%'], color='red', alpha=0.18, s=abs(y['comp_Total'])**1.8)

legend1 = ax.legend(['Win', 'Loss'],
                    loc="upper left", title="Win", prop={'size': 19})
ax.add_artist(legend1)

plt.axvline(x=0, color='black', linestyle='--')
plt.axhline(y=0, color='black', linestyle='--')

# produce a legend with the unique colors from the scatter

# produce a legend with a cross section of sizes from the scatter
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = ax.legend(handles, range(1, x['comp_Total'].max()), loc="lower right", title="Win margin",
                    bbox_to_anchor=(1.22, 0), prop={'size': 25})

plt.title("Comparative 3-Point Percentage vs Comparative 3-Points Made 2018")
plt.ylabel("Comparative 3-Point Percentage")
plt.xlabel("Comparative 3-Points Made")
plt.savefig('../../../graphs/3000_2018_made_vs_percentage_win_loss.png',bbox_inches='tight')
plt.show()
```

# Number of attempts vs percentage

## 2018

```python
fig, (ax1, ax2,ax3) = plt.subplots(3, sharex=True,figsize=(10,12))
fig.suptitle('3 Point Percentage vs 3 Points Attempted vs 3 Points Made 2018', fontsize=24)
fig.tight_layout()
fig.subplots_adjust(top=0.93)
ax1.set_title('3 Point Percentage vs Win Margin')
ax2.set_title('3 Points Attempted vs Win Margin')
ax3.set_title('3 Points Made vs Win Margin')
ax1.axes.xaxis.set_visible(False)
ax2.axes.xaxis.set_visible(False)
x = df[(df['Season'] == 2018)]

ax1.scatter(x['comp_Total'], x['comp_3P%'], alpha=0.4)
m1, b1 = np.polyfit(x['comp_Total'], x['comp_3P%'], 1)
corr = ax1.plot(x['comp_Total'], m1*x['comp_Total'] + b1, color='black', linestyle='--')
ax1.set_ylabel('Comparative Percentage')
fig.legend(corr, ['Regression'], loc='lower right', bbox_to_anchor=(1.25, 0.9))

ax2.scatter(x['comp_Total'], x['comp_3PA'], alpha=0.4)
m2, b2 = np.polyfit(x['comp_Total'], x['comp_3PA'], 1)
ax2.plot(x['comp_Total'], m2*x['comp_Total'] + b2, color='black', linestyle='--')
ax2.set_ylabel('Comparative Attempted')

ax3.scatter(x['comp_Total'], x['comp_3PM'], alpha=0.4)
m3, b3 = np.polyfit(x['comp_Total'], x['comp_3PM'], 1)
ax3.plot(x['comp_Total'], m3*x['comp_Total'] + b3, color='black', linestyle='--')
ax3.set_ylabel('Comparative Made')
ax3.set_xlabel('Win Margin')

line1 = round(x['comp_Total'].corr(x['comp_3P%']), 3)
plt.text(50,103, line1, fontsize=16)

line2 = round(x['comp_Total'].corr(x['comp_3PA']), 3)
plt.text(50, 50, line2, fontsize=16)

line3 = round(x['comp_Total'].corr(x['comp_3PM']), 3)
plt.text(50, 5, line3, fontsize=16)

plt.savefig('../../../graphs/3000_attempts_vs_percentage_2018.png',bbox_inches='tight')
plt.show()
```
