```python
t1 = conf.pivot(index='TEAM_NAME', columns='SEASON', values='OER')[[2004,2019]]
t2 = east.pivot(index='TEAM_NAME', columns='SEASON', values='OER')[[2004,2019]]
```

```python
plt.figure(figsize=(5,5))
ax = pd.plotting.parallel_coordinates(t1.reset_index(), 'TEAM_NAME', color='orange', alpha=0.7)

pd.plotting.parallel_coordinates(
        t2.reset_index(), 'TEAM_NAME', ax=ax, 
        axvlines_kwds=dict(color='grey', lw=2/2), 
        color='purple', alpha=1
    )
import matplotlib.patches as mpatches
east = mpatches.Patch(color='orange', label='East')
west = mpatches.Patch(color='purple', label='West')
ax.legend(handles=[east, west])
ax.grid(False)
ax.set_title('OER 2003 to 2019')
```

```python
plt.figure(figsize=(5,5))
plt.tight_layout()
ax = pd.plotting.parallel_coordinates(t1.reset_index(), 'TEAM_NAME', color='blue', alpha=0.5)

pd.plotting.parallel_coordinates(
        t2.reset_index(), 'TEAM_NAME', ax=ax, 
        axvlines_kwds=dict(color='grey', lw=2/2), 
        color='orange', alpha=0.5
)
east = mpatches.Patch(color='orange', label='East')
west = mpatches.Patch(color='blue', label='West')
ax.legend(handles=[east, west])
ax.set_ylabel("Offensive Rating")
ax.set_title('OER 2004 to 2019')
plt.savefig('C:/Users/User/OneDrive/Desktop/DSIP/NBAproject/graphs/1100_OER_2004_2019.png',bbox_inches='tight');
```
