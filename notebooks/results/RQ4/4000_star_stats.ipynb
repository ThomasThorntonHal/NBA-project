# Introduction
In this notebook we will look at how the stats of star players have changed over the years.

# Notebook Objectives
We are interested in three questions:
1. How has the efficiency of top players changed?
2. How has the usage percentage of top players changed?
3. What stast are valued in an MVP?

# Top players usage percentage per season

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8),
      (ax9, ax10, ax11, ax12), (ax13, ax14, ax15, ax16)) = plt.subplots(4, 4, figsize=(22, 14), sharex=True, sharey=True)
fig.suptitle('EFF vs USG% top 30 players by season', fontsize=24)

count=0
for ax in fig.get_axes():
    x, y = graph_df.loc[2004+count]['EFF'], graph_df.loc[2004+count]['USG%']
    ax.scatter(x, y, alpha=0.3)
    m1, b1 = np.polyfit(x, y, 1)
    ax.plot(x, m1*x + b1, color='black', linestyle='--')
    ax.set_title(str(2004+count))
    count = count + 1
    ax.label_outer()

fig.text(0.51, 0.06, 'Player Efficiency', ha='center', fontsize=24)
fig.text(0.08, 0.5, 'Usage percentage', va='center', rotation='vertical', fontsize=24)
plt.savefig('../../../graphs/4000_EFF_vs_USG_top_players.png',bbox_inches='tight');
```

List of MVPs and the year they won

```python
MVP = {2003:'Kevin Garnett', 2004:'Steve Nash', 2005:'Steve Nash', 2006:'Dirk Nowitzki',
       2007:'Kobe Bryant', 2008:'LeBron James', 2009:'LeBron James', 2010:'Derrick Rose', 2011:'LeBron James',
       2012:'LeBron James', 2013:'Kevin Durant', 2014:'Stephen Curry', 2015:'Stephen Curry', 2016:'Russell Westbrook',
       2017:'James Harden', 2018:'Giannis Antetokounmpo', 2019:'Giannis Antetokounmpo'}
```

```python
# Create and empty dataframe
mvp_df = tmp_df[tmp_df['EFF']==1242].copy()
# Fill the dataframe with MVP players
for player in MVP:
    mvp_df = mvp_df.append(tmp_df.loc[player].reset_index().set_index(['PLAYER_NAME']).loc[MVP.get(player)])
```

```python
mvp_df = mvp_df.reset_index().set_index(['SEASON_TEAM'])
mvp_df
```

List of rookies of the year

```python
ROTY = {2003:'LeBron James', 2004:'Emeka Okafor', 2005:'Chris Paul', 2006:'Brandon Roy', 2007:'Kevin Durant',
        2008:'Derrick Rose', 2009:'Tyreke Evans', 2010:'Blake Griffin', 2011:'Kyrie Irving', 2012:'Damian Lillard',
        2013:'Michael Carter-Williams', 2014:'Andrew Wiggins', 2015:'Karl-Anthony Towns', 2016:'Malcolm Brogdon',
        2017:'Ben Simmons', 2018:'Luka Doncic', 2019:'Ja Morant'}
```

```python
# Make an empty dataframe
roty_df = tmp_df[tmp_df['EFF']==1242].copy()
# Add rookie of the year data
for player in ROTY:
    roty_df = roty_df.append(tmp_df.loc[player].reset_index().set_index(['PLAYER_NAME']).loc[ROTY.get(player)])
```

```python
roty_df
```

Because some players have won more than one MVP we need to make a list of the unique MVPs

```python
MVP_players = []
for player in MVP:
    if MVP.get(player) not in MVP_players:
        MVP_players.append(MVP.get(player))
```

```python
MVP_players
```

```python
MVPs = seasonal_df.reset_index()
MVPs
```

Filter the season data to only include MVP seasons

```python
MVPs = MVPs[MVPs['PLAYER_NAME'].isin(MVP_players)]
MVPs.head()
```

Make a unique dataframe for each MVP

```python
KG = MVPs[MVPs['PLAYER_NAME']=='Kevin Garnett'].set_index(['SEASON'])
NASH = MVPs[MVPs['PLAYER_NAME']=='Steve Nash'].set_index(['SEASON'])
DIRK = MVPs[MVPs['PLAYER_NAME']=='Dirk Nowitzki'].set_index(['SEASON'])
KOBE = MVPs[MVPs['PLAYER_NAME']=='Kobe Bryant'].set_index(['SEASON'])
LBJ = MVPs[MVPs['PLAYER_NAME']=='LeBron James'].set_index(['SEASON'])
DROSE = MVPs[MVPs['PLAYER_NAME']=='Derrick Rose'].set_index(['SEASON'])
KD = MVPs[MVPs['PLAYER_NAME']=='Kevin Durant'].set_index(['SEASON'])
CURRY =  MVPs[MVPs['PLAYER_NAME']=='Stephen Curry'].set_index(['SEASON'])
WESTBROOK =  MVPs[MVPs['PLAYER_NAME']=='Russell Westbrook'].set_index(['SEASON'])
HARDEN =  MVPs[MVPs['PLAYER_NAME']=='James Harden'].set_index(['SEASON'])
GIANNIS =  MVPs[MVPs['PLAYER_NAME']=='Giannis Antetokounmpo'].set_index(['SEASON'])
```

# MVPs vs Rookies

We will use this dataframe to get the player with the max efficiency for each season

```python
max_per_season = df.groupby(['SEASON', 'PLAYER_ID', 'PLAYER_NAME']).mean().copy()
max_per_season
```

```python
max_eff_per_season = {}
for season in range(2003,2020):
    max_eff_per_season[season] = max_per_season.loc[season]['EFF'].max()
max_eff_per_season
```

```python
plt.figure(figsize=(18, 9))
ax = plt.subplot()
plt.title('NBA MVPs vs ROTYs vs League', fontsize=24)

ax.set_xlabel('Season', fontsize=20)
ax.set_ylabel('Efficiency', fontsize=20)

legend_elements = [Line2D([0], [0], color='black', lw=4, label='Most efficient player'),
                   Line2D([0], [0], color='blue', lw=4, label='Most valuable player'),
                   Line2D([0], [0], color='green', lw=4, label='Rookie of the year'),
                   Line2D([0], [0], color='red', lw=4, label='League average')]
ax.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.25, 1))

# league average
ax.plot(df.groupby(['SEASON']).mean()['EFF'], color='red', linewidth=4)

ax.plot(mvp_df.reset_index().set_index(['SEASON_TEAM'])['EFF'], color='blue')

ax.plot(roty_df.reset_index().set_index(['SEASON_TEAM'])['EFF'], color='green')

seasons = list(max_eff_per_season.keys())
eff = list(max_eff_per_season.values())
ax.plot(seasons, eff, color='black')

ax.set_xlim(2003,2019)

plt.savefig('../../../graphs/4000_MVP_vs_ROTY_vs_EFF.png',bbox_inches='tight')
plt.show()
```
