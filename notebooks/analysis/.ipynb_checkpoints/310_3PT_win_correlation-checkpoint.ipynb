# Introduction
In this notebook I will merge the 3 point dataset with itself to add the opposing team's 3 point data. Then I will remove the unnecessary rows. Usin these new columns I can compare the number of 3 point shots the team scored in comparison to their opposition. I am doing this because if one team scores 30 3 point shots and loses but their opposition scores 40 3 point shots, this is different from a team scoring 30 3 point shots and their opposition scoring 0 and still losing.

## Notebook Objective
Add opposition columns and calculate new columns that comapres first team and second team.

# Setup

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import urllib.request, json
from unicodedata import normalize
import os 
import sys
```

# Parameters

```python
# In
GAMES_DATASET = '../../data/processed/003_clean_3point.csv'

# Out
OPPOSING_TEAM_SHOTS = '../../data/processed/310_3point_opp.csv'
```

# Load Dataset 

```python
games_df = pd.read_csv(GAMES_DATASET)
games_df.shape, games_df.dtypes
```

```python
games_df.head()
```

Add the opposing teams 3 pointer data.

```python
merged = pd.merge(games_df, games_df[['Game', 'Total', '3PM_team', '3PA_team', '3P%_team', 'Team']],
                  how='left', left_on=['Game'], right_on = ['Game'], suffixes=(None, '_OPP'))
merged.head()
```

```python
merged.columns
```

Now we need to drop duplicate columns.

```python
merged = merged[merged['Team'] != merged['Team_OPP']].copy()
merged.head()
```

## Calculate comparison
Now that we have the opposition's 3 point data we can calculate the difference between the performance of the two teams.

```python
merged['comp_3PM'] = merged['3PM_team']-merged['3PM_team_OPP']
merged['comp_3PA'] = merged['3PA_team']-merged['3PA_team_OPP']
merged['comp_3P%'] = merged['3P%_team']-merged['3P%_team_OPP']
merged['comp_Total'] = merged['Total']-merged['Total_OPP']
```

```python
merged.head()
```

## Visualising the data
I want to make some quick visualisations to see what the data looks like

```python
x = merged[(merged['Season'] == 2015) & (merged['Team'] == 'GSW')]['comp_3P%']
y = merged[(merged['Season'] == 2015) & (merged['Team'] == 'GSW')]['comp_Total']

plt.plot(x, y, 'o');
```

```python
plt.figure(figsize=(12,11))
x = merged[(merged['Season'] == 2018) & (merged['Won'] == 1)]
y = merged[(merged['Season'] == 2018) & (merged['Won'] == 0)]
plt.scatter(x['comp_3PM'], x['comp_3P%'], color='green', alpha=0.4, s=6*abs(x['comp_Total']))
plt.scatter(y['comp_3PM'], y['comp_3P%'], color='red', alpha=0.4, s=6*abs(y['comp_Total']))
plt.show()
```

This is a funciton to correct the columns that contain percentages after a groupby operation.

```python
def correct_percent(df):
    df['FG%_team'] = df['FGM_team']/df['FGA_team']
    df['3P%_team'] = df['3PM_team']/df['3PA_team']
    df['FT%_team'] = df['FTM_team']/df['FTA_team']
```

```python
season18 = games_df[games_df['Season'] == 2018].groupby(['Team']).mean()
season18.head()
```

```python
correct_percent(season18)
season18.head()
```

```python
games_df[games_df['Season'] == 2015].groupby(['Team']).mean().sort_values('3P%_team', ascending=False)[['3P%_team', 'Won']]\
                                    .plot(kind='bar', figsize=(18,5), title='2015 win% and 3p%');
```

```python
games_df[games_df['Season'] == 2015].groupby(['Team']).mean().sort_values('3P%_team', ascending=False)[['3P%_team', 'Won']]
```

```python
fig = plt.figure(figsize=(18,5)) # Create matplotlib figure

ax = fig.add_subplot() # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.4

wins = games_df[games_df['Season'] == 2015].groupby(['Team']).mean().sort_values('3P%_team', ascending=False)[['Won']]
threes = games_df[games_df['Season'] == 2015].groupby(['Team']).mean().sort_values('3P%_team', ascending=False)[['3P%_team']]

wins.plot(kind='bar', color='orange', title='2015 3P% against win%', ax=ax, width=width, position=1)
threes.plot(kind='bar', color='lightgrey', ax=ax2, width=width, position=0)

ax.set_ylabel('Win%')
ax2.set_ylabel('3P%')
ax2.set_ylim([0.27, 0.39])
fig.legend()

plt.show()
```

# Sava Dataset
Now that we have added opponent's data and created comparison columns we can save the new dataset.

```python
merged.to_csv(OPPOSING_TEAM_SHOTS, index=False)
merged.columns, OPPOSING_TEAM_SHOTS
```
